## Expected Reward

$$S_i=\sum_{i=0}^{i \to \infty}P_N^i·P_Z·i·\text{AVG}$$

$$S_n=\sum_0^n\left(a+k\cdot b\right)r^k$$
$$r\cdot S_n=\sum_0^n\left(a+k\cdot b\right)r^{k+1}$$

$$S_n-r\cdot S_n=\left(1-r\right)S_n=a+b\sum_{k=1}^nr^{k+1}-\left(a+n\cdot b\right)r^{n+1}$$

Here we can solve for <b>Sn</b>, after summing the geometric progression. If <b>|r| &lt; 1</b>, the last term tends to <b>0</b> when <b>n  → ∞</b>, resulting

$$\boxed{S=\frac{a}{1-r}+\frac{b\cdot r}{\left(1-r\right)^2}=\frac{a+\left(b-a\right)r}{\left(1-r\right)^2}}$$

In our example, $a = 0, b = AVG·P_Z,  r = P_N$.

$$E = \frac{AVG·P_Z·P_N}{(1-P_N)^2}$$

Expected Reward:
$$E = Rewards·\frac{n_z}{(n_z+n_k)^2}$$

$$\boxed{E = TR·\frac{Z}{(Z + K)^2}}$$

## Expected "episode" length
$$E_0 = 1·P_K+1·P_Z+P_N·(1+E_0) \implies E_0 = \frac{1}{1-P_N} = 1+\frac{n_v}{n_z+n_k}$$
